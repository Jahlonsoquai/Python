```python
#Script Juego del Ahorcado:
import time
import random

marcas_autos=["ford", "audi", "nissan", "renault", "chevrolet", "mitsubishi", "cadillac", "suzuki"]
nombres_paises = ["mexico", "colombia", "canada", "venezuela", "uruguay", "argentina", "guatemala", "panama", "ecuador"]

nombre = input("¿Cuál es tu nombre?\n")

time.sleep(1.5)
print("\n"+"="*50)
print(f'HOLA {nombre.upper()} BIENVENIDO(A) AL JUEGO DEL AHORCADO')
print("="*50,"\n")

time.sleep(1.5)
print("El objetivo es adivinar la palabra secreta, letra por letra.")
print("Tienes 7 vidas, pierdes una vida cada que te equivocas.\n")

time.sleep(1.5)
print(f'''Opciones de grupo de palabras para el juego:
1. Marcas de automóviles.
2. Nombres de países.
''')
#Aquí se códifica el menú de opciones.
time.sleep(1)
opcion_selecta = int(input("Ingrese la opción deseada: "))

while True:
    
    if opcion_selecta == 1:
        print("¡Bien! Jugaras con marcas de automóviles.\n")
        secretWord = random.choice(marcas_autos) #seleccionara de forma aleatoria una palabra(string) de la colección marcas_autos y la designará como palabra secreta.
        break 
    elif opcion_selecta == 2:
        print("¡Bien! Jugaras con nombres de países.\n")
        secretWord = random.choice(nombres_paises)
        break
    else:
        print("Por Favor selecciona una opción válida.\n")
        opcion_selecta = int(input("Ingrese la opción deseada: "))

patibulo = [
'''  
+______+ 
|       
|       
|        
|        
|        
|        
+========

''',
'''  
+______+ 
|      | 
|      | 
|        
|        
|        
|        
+========

''',
'''  
+______+ 
|      | 
|      | 
|      O  
|        
|        
|        
+========

''',
'''
+______+ 
|      | 
|      | 
|      O  
|      |  
|        
|        
+========

''',
'''
+______+ 
|      | 
|      | 
|      O  
|     /|  
|        
|        
+========

''',
''' 
+______+ 
|      | 
|      | 
|      O  
|     /|  
|     /   
|        
+========

''',
'''  
+______+ 
|      | 
|      | 
|      O  
|     /|)  
|     /   
|        
+========

''',
'''
+______+ 
|      | 
|      | 
|      O  
|     /|)  
|     / )  
|
+========
>>AHORCADO  

''']       

ahorcado = {"vida7":patibulo[0], "vida6":patibulo[1],"vida5":patibulo[2],"vida4":patibulo[3], "vida3":patibulo[4], "vida2":patibulo[5], "vida1":patibulo[6],"vida0":patibulo[7]}

print("¡Adivina la palabra secreta! de ",(len(secretWord))," letras\n")
#print("".join(lis_guessed),'\n')

letterIn = [] #Es una lista para verificar las letras que el usuario va ingresando.
lifes = 7
s = "-"

lis_guessed = [] # Se crea está lista para desagregar la palabra secreta.
for i in range(len(secretWord)):
  lis_guessed.append('_ ') #Se agrega este string con la longitud de la palabra a la lista.

usadas = len(secretWord)  #Aquí establecemos lo que mide la palabra secreta para entrar al while

while usadas > 0 and lifes > 0: #usadas vale la longitud de la palabra.
 
  if lifes == 7:
    print(ahorcado.get("vida7"))
  elif lifes == 6:
    print(ahorcado.get("vida6"))
  elif lifes == 5:
    print(ahorcado.get("vida5"))
  elif lifes == 4:
    print(ahorcado.get("vida4"))
  elif lifes == 3:
    print(ahorcado.get("vida3"))
  elif lifes == 2:
    print(ahorcado.get("vida2"))
  elif lifes == 1:
    print(ahorcado.get("vida1"))
        

  print(f'Tienes {lifes} vidas y las letras que has ingresado son: {s.join(letterIn)}')
  print("".join(lis_guessed),'\n') #Se manda a imprimir un string y se une con la palabra listada, y se mostrará de forma incógnita.
  userChar = input("Ingresa una letra:\n")
         
  for char in range(len(secretWord)):
      if userChar == secretWord[char]:
        lis_guessed[char] = userChar.upper() + " "
    
  if userChar in letterIn:
    print("Esa letra ya se utilizó, ingresa otra letra")
  else:
    letterIn.append(userChar)
    
    if userChar in secretWord:
      #Sirve para determinar la primera condición del while.
      usadas = usadas - secretWord.count(userChar) #aqui restamos a la longitud inicial el número de veces que aparece la palabra y la almacenamos en una variable(que es la misma pero es necesario para que respete el valor)
      #print(usadas)
      print(f'La letra {userChar}, si está {secretWord.count(userChar)} veces en la palabra secreta')
          
    else:   
      lifes -= 1
      print(f'La letra {userChar}, no se encuentra en la palabra secreta')

if lifes == 0:
  print(f'\nYa no te quedan más oportunidades.\n¡Lastima!\nLa palabra secreta era: {secretWord.upper()}')
  print(ahorcado.get("vida0"))
else:
  print("\n¡Felicidades Adivinaste la Palabra! "," ".join(secretWord.upper()))



```
